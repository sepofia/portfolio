*Task 1: Comprehensions and operations on sets*

Perform the following transformations (each action is solved in one line). After each action, display the result on the screen.

1. Generate a list of all powers of two from 1 to 200 (the generated values must be in the range from 1 to 200). Start with the 1st degree: 2, 4, 8 and so on. Put the result in `power_2_numbers`
2. Generate a set of unique values from 8 to 50 inclusive. If the element is divisible by 5, replace it with one, and leave the rest as it is. Put the result in the `generated_set'. The type of the variable should be - `set`
3. Create a set type variable named `new_set`, which will contain all the elements from `power_2_numbers` and `generated_set`, except for the elements that occur in both 'power_2_numbers' and `generated_set'.
4. Remove from the `new_set` all elements that are less than 34 or more than 120.
4. Generate a dictionary `unicode_values`, in which the key will be all the numbers from `new_set`, and the values will be the corresponding Unicode characters.

To get a character by its Unicode value, you need to call the `chr()` function. In parentheses, you need to pass the number by which you want to get the character.

```python
# some examples
code = ord('u')
print(f'Unicode-value for u: {code}')
symbol = chr(code)
print(f'Character for unicode-value {code}: {symbol}')
```

```python
power_2_numbers = []
pow = 2
while pow < 200:
    power_2_numbers.append(pow)
    pow *= 2

print(power_2_numbers)
```

```python
generated_set = set()
for i in range(8, 51):
    num = 1 if i % 5 == 0 else i
    generated_set.add(num)

print(generated_set)
```

```python
new_set = set(power_2_numbers) ^ generated_set
print(new_set)
```

```python
bad_elem = []
for elem in new_set:
    if elem > 120 or elem < 34:
        bad_elem.append(elem)

for elem in bad_elem:
    new_set.discard(elem)

print(new_set)
```

```python
unicode_values = {elem: chr(elem) for elem in new_set}

print(unicode_values)
```

*Task 2. Duty*

The dictionary `schedule` is given, which stores information about shifts. The key is the employee id, the value is a list of the days on which this employee can go on duty. The days' names are listed in the `unique_days` list.

Find it:

1. Days when no one can be on duty (use sets to solve this point).
2. For each day of the week, find the number of employees who can go on duty on the corresponding day.

Use functions and decomposition.

```python
shedule = {
    '1001': ['Mo', 'Su'],
    '1231': ['Mo', 'Tu'],
    '1232': ['We', 'Th', 'Fr'],
    '1280': ['We', 'Fr', 'Su'],
    '1282': ['Th'],
    '1290': ['Fr', 'Su'],
    '1303': ['Tu', 'Su'],
}
```

```python
unique_days = ['Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa', 'Su']


def emply_days(shedule: dict) -> None:
    week = set(unique_days)
    days = set()
    for employee in shedule:
        for day in shedule[employee]:
            days.add(day)
    print(f'Days when no one can be on duty: {week - days}')


def count_employees(shedule: dict) -> None:
    count_empl = {day: 0 for day in unique_days}
    for employee in shedule:
        for day in shedule[employee]:
            count_empl[day] += 1
    print(f'Number of employees for each day: {count_empl}')


def analyse_shedule(shedule: dict) -> None:
    emply_days(shedule)
    count_employees(shedule)
```

```python
analyse_shedule(shedule)
```

*Task 3. The distance between two points*

The list of `coords` contains 2 tuples, each of which is the coordinates of points A and B, respectively. Write a function that looks for the distance between these points.

```python
coords1 = [(1, 10), (1, 11)]
coords2 = [(0, 20), (20, 20)]
coords3 = [(100, 91, 87), (80, 35, 70)]
coords4 = [(0, 100, 20), (0, 0, 80)]
```

```python
def distance(coords: list[tuple[float]]) -> float:
    if len(coords) != 2:
        return "Coordinates for two points are necessary"
    return (sum([(a - b) ** 2 for a, b in zip(coords[0], coords[1])])) ** 0.5
```

```python
for coords in [coords1, coords2, coords3, coords4]:
    print(distance(coords))
```

*Task 4. Accounting of goods*

Two lists are given: `ids`, `count'. The first list contains the unique identifiers of the goods, the second list contains the number of corresponding goods in stock.

From the data in the lists, create a dictionary `items_count`, which will store the quantity of goods in stock. The key is the product ID, the value is the quantity. Make the decision in one line.

The product ID is a string of the form: `number1:number2`. Number1 indicates the product category, and number2 indicates its number. Calculate the following statistics on this data:
1. The number of unique categories (formalize the solution in one line).
2. The average number of **units of goods** in stock in the category (round to 1 digit after the decimal point).
3. The average number of **unique products** in the category.
4. ID of the items with the maximum number of units in stock.

```python
ids = ['1111:271', '1111:190', '1231:106', '1211:104', '1111:201', '1231:120', '1001:205', '1001:223', '1001:127', '1001:236', \
      '1111:229', '1231:286', '1231:195', '1001:279', '1001:124', '1111:292', '1505:219', '1231:259', '1231:253', '1001:220', '1001:202',\
      '1231:103', '1211:249', '1212:275']
count = [111, 3, 13, 111, 9, 5, 17, 10, 13, 3, 16, 4, 16, 11, 18, 12, 14, 4, 3, 2, 14, 14, 10, 10]
```

```python
items_count = {id_i: count_i for id_i, count_i in zip(ids, count)}

print(items_count)
```

```python
# auxiliary dictionary:
# key - category, value - list of pairs [item number, quantity]
cathegories = {}
for id_i, count_i in zip(ids, count):
    cathegory_i, item = id_i.split(':')
    if cathegory_i in cathegories:
        cathegories[cathegory_i][0].append(item)
        cathegories[cathegory_i][1].append(count_i)
    else:
        cathegories[cathegory_i] = [[item], [count_i]]


# point 1
count_unique_cathegories = len(set([id_i.split(':')[0] for id_i in ids]))
print(f'Number of unique categories: {count_unique_cathegories}')

# point 2
avg_count = {key: round(sum(value) / len(value), 1) for (key, [_, value]) in cathegories.items()}
print(f'The average number of item in the stock for each category: {avg_count}')

# point 3
avg_count_of_items = sum([len(value) for (_, [value, _]) in cathegories.items()]) / count_unique_cathegories
print(f'The average number of unique items for each category: {avg_count_of_items}')

# point 4
max_ids = {}
for cath_i in cathegories:
    max_count, max_id = 0, 0
    items, counts = cathegories[cath_i]
    for item_i, count_i in zip(items, counts):
        if count_i > max_count:
            max_count, max_id = count_i, item_i
    max_ids[cath_i] = max_id
print(f"Item's ID with maximum number of units in stock for each category: {max_ids}")
```

*Task 5. Word frequency — part 1*

Often, when building machine learning models based on texts, one of the input parameters of the model is the frequency of occurrence of a word in the text. In this problem, we will solve the problem of finding the frequency of a word in a document.

The `document_text` variable contains the text. You need to find the frequency of use of each unique word and put it in the dictionary `word_frequency'. There is no need to convert words to the initial form.

Recommendations:
1. To remove unnecessary characters, use the `replace` method for texts.
2. To count the frequencies, use the `count` method on the lists.

```python
document_text = 'There are the Food Courts in the cities that seats a lot of people and caters for every taste with dishes from all around the world. \
 The aromatic smells that come from the Food Courts arc so delicious that your mouth will water. \
 In recent years foreign foods have become a regular part of our life.'
```

```python
def word_frequency(text: str) -> dict[str, int]:
    text = text.lower()
    for symbol in '.,:;!?-':
        text = text.replace(symbol, ' ')
    words = text.split()
    frequency = {}
    for word in words:
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    return frequency
```

```python
freq_dict = word_frequency(document_text)
```

*Task 7. The area of the triangle*

The list of `coords` contains 3 tuples, each of which corresponds to the coordinate of the vertex of an isosceles triangle on the coordinate plane. Write a function that finds the area of an isosceles triangle by its coordinates.

The area of the triangle can be found using the following formula:

> S = 0.5 * h * (A1A2), where

> A1A2 — distance from point A1 to point A2;

> h — height (distance from point A3 to the middle of the segment A1A2).

```python
coords = [(3, 2), (11, 2), (7, 10)]
```

```python
def point_order(coords: list[tuple[float]]) -> list[tuple[float]]:
    side_1 = distance([coords[0], coords[1]])
    side_2 = distance([coords[1], coords[2]])
    side_3 = distance([coords[2], coords[0]])
    if side_1 == side_2:
        return coords
    if side_2 == side_3:
        return [coords[1], coords[2], coords[0]]
    if side_1 == side_3:
        return [coords[2], coords[0], coords[1]]


def segments(coords: list[tuple[float]]) -> tuple[float]:
    a = distance([coords[0], coords[2]])
    center_point = [(x1 + x2) / 2 for x1, x2 in zip(coords[0], coords[2])]
    h = distance([coords[1], center_point])
    return a, h



def square(coords: list[tuple[float]]) -> float:
    if len(coords) != 3:
        return "необходимы координаты трёх точек"

    a, h = segments(point_order(coords))
    return 0.5 * a * h
```
